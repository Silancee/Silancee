```python
# Gerekli Kütüphaneleri İçe Aktarma
import tensorflow as tf
from tensorflow_datasets import load
from tensorflow.data import AUTOTUNE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau,
    ModelCheckpoint
)
import matplotlib.pyplot as plt
import numpy as np

# --- 1. VERİ SETİNİ YÜKLEME VE HAZIRLAMA ---

IMAGE_SIZE = (224, 224)
# BELLEK SORUNUNU ÇÖZMEK İÇİN BATCH BOYUTU KÜÇÜLTÜLDÜ (32 -> 16)
BATCH_SIZE = 16

print("Veri Seti Yükleniyor...")

# 'tf_flowers' veri setini yükle
(train_ds, val_ds, test_ds), metadata = load(
    'tf_flowers',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True
)

NUM_CLASSES = metadata.features['label'].num_classes
print(f"Toplam Sınıf Sayısı: {NUM_CLASSES}")

# Veri Ön İşleme Fonksiyonu
def preprocess(image, label):
    # Görüntüleri belirlenen boyuta yeniden boyutlandır (224x224)
    image = tf.image.resize(image, IMAGE_SIZE)
    # Piksel değerlerini [0, 1] aralığına normalleştir
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Veri Seti Optimizasyonu ve Ön İşleme
train_ds = train_ds.map(preprocess, num_parallel_calls=AUTOTUNE) \
                   .cache() \
                   .shuffle(buffer_size=1000) \
                   .batch(BATCH_SIZE) \
                   .prefetch(buffer_size=AUTOTUNE)

val_ds = val_ds.map(preprocess, num_parallel_calls=AUTOTUNE) \
               .batch(BATCH_SIZE) \
               .prefetch(buffer_size=AUTOTUNE)

test_ds = test_ds.map(preprocess, num_parallel_calls=AUTOTUNE) \
                .batch(BATCH_SIZE) \
                .prefetch(buffer_size=AUTOTUNE)

print("Veri setleri hazırlandı.")

# --- 2. CNN MODELİNİ TANIMLAMA ---

def create_cnn_model():
    model = Sequential([
        # İlk Evrişim Katmanı
        Conv2D(32, (3, 3), activation='relu', input_shape=IMAGE_SIZE + (3,)),
        MaxPooling2D((2, 2)),

        # İkinci Evrişim Katmanı
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),

        # Üçüncü Evrişim Katmanı
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),

        # Yoğun Katmanlara Geçiş
        Flatten(),
        Dropout(0.5),
        Dense(256, activation='relu'),

        # Çıkış Katmanı
        Dense(NUM_CLASSES, activation='softmax')
    ])

    # Modeli derleme
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

model = create_cnn_model()
model.summary()

# --- 3. CALLBACK'LERİ TANIMLAMA ve MODEL EĞİTİMİ ---

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
lr_plateau = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
model_checkpoint = ModelCheckpoint(
    'best_flower_cnn_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

EPOCHS = 50

print("\nModel Eğitimi Başlıyor...")

history = model.fit(
    train_ds,
    epochs=EPOCHS,
    validation_data=val_ds,
    callbacks=[early_stopping, lr_plateau, model_checkpoint]
)

# --- 4. MODELİ DEĞERLENDİRME VE GÖRSELLEŞTİRME ---

print("\nModel Test Ediliyor...")

loss, accuracy = model.evaluate(test_ds)
print(f"Test Kaybı: {loss:.4f}")
print(f"Test Doğruluğu: {accuracy*100:.2f}%")

# Eğitim Geçmişini Çizme
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Doğruluk Eğrisi')
plt.xlabel('Epok')
plt.ylabel('Doğruluk')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Kayıp Eğrisi')
plt.xlabel('Epok')
plt.ylabel('Kayıp')
plt.legend()

plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.image import resize
import numpy as np
import matplotlib.pyplot as plt
# Colab'de dosya yükleme arayüzünü açmak için gereklidir
from google.colab import files

# --- 1. SABİTLER, SINIF ADLARI ve BİLGİ SÖZLÜĞÜ TANIMLAMA ---

IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['dandelion', 'daisy', 'tulips', 'sunflowers', 'roses']

# Tahmin edilen çiçekler hakkında bilgi içeren sözlük
FLOWER_INFO = {
    'dandelion': "Karahindiba (Dandelion): Genellikle baharın müjdecisi olarak kabul edilir ve yenilebilir bir yabani ottur. Yaprakları salatalarda kullanılır. Sağlık açısından faydalıdır.",
    'daisy': "Papatya (Daisy): Saflığın ve masumiyetin sembolüdür. Genellikle 'seviyor sevmiyor' falında kullanılır. Çay olarak tüketildiğinde sakinleştirici etkisi vardır.",
    'tulips': "Lale (Tulip): Hollanda ile özdeşleşse de, anavatanı Türkiye'dir. Geçmişte Osmanlı İmparatorluğu'nda büyük bir kültürel ve ekonomik değere sahipti. Soğanlı bir bitkidir.",
    'sunflowers': "Ayçiçeği (Sunflower): Güneşi takip etme özelliğiyle bilinir (heliotropizm). Tohumları yağ ve çerez yapımında yaygın olarak kullanılır. Dünyanın en büyük çiçeklerindendir.",
    'roses': "Gül (Rose): Aşkın, güzelliğin ve tutkunun evrensel sembolüdür. Rengine göre farklı anlamlar taşır. Parfümeri ve kozmetik sektöründe sıklıkla kullanılır."
}

# --- 2. YARDIMCI FONKSİYONLAR ---

def load_and_preprocess_image(img_path):
    # 1. Resmi yükle
    img = tf.io.read_file(img_path)
    # 2. Resmi tensöre dönüştür (RGB formatında)
    img = tf.image.decode_image(img, channels=3)
    # 3. Boyutu modelin beklediği boyuta getir (224x224)
    img = resize(img, IMAGE_SIZE)
    # 4. Normalleştir ([0, 1] aralığına getir)
    img = tf.cast(img, tf.float32) / 255.0
    # 5. Modeli tahmin için hazırla (Batch boyutunu ekle: (1, 224, 224, 3))
    img = tf.expand_dims(img, axis=0)
    return img

def display_flower_info(predicted_name):
    # Sözlükten ilgili bilgiyi al ve yazdır
    info = FLOWER_INFO.get(predicted_name)
    if info:
        print("\n🌸 Çiçek Hakkında Bilgi:")
        # Bilgiyi daha okunur hale getirmek için yeni satır karakteri ekliyoruz
        print(f"**{predicted_name.upper()}**: {info}")
    else:
        print("\nBu çiçek türü hakkında bilgi bulunamadı.")

# --- 3. ANA ÇALIŞTIRMA BLOĞU ---

# Kullanıcıdan dosya yüklemesini iste
print("Lütfen tahmin etmek istediğiniz çiçek resmini yükleyin.")
uploaded = files.upload()

# Yüklenen resmin dosya yolunu al
# files.upload() genellikle bir sözlük döndürür, ilk anahtarı alıyoruz
image_path = list(uploaded.keys())[0]

# 1. Kaydedilmiş Modeli Yükle
try:
    # Modelin ilk kod bloğunda kaydedilmiş olması gerekir
    loaded_model = load_model('best_flower_cnn_model.h5')
except Exception as e:
    print("\n" + "="*60)
    print("HATA: Model yüklenemedi. Lütfen önce BATCH_SIZE=16 olan eğitim kodunu çalıştırın.")
    print(f"Hata detayı: {e}")
    print("="*60)
    exit()

# 2. Resmi Ön İşleme Tabi Tut
processed_image = load_and_preprocess_image(image_path)

# 3. Tahmin Yap
predictions = loaded_model.predict(processed_image)

# 4. En Yüksek Olasılıklı Sınıfı Bul
predicted_class_index = np.argmax(predictions[0])
predicted_class_name = CLASS_NAMES[predicted_class_index]
confidence = np.max(predictions[0]) * 100

# 5. Sonucu Yazdır ve Görselleştir
print("\n" + "="*30)
print("     💐 TAHMİN SONUCU")
print("="*30)
print(f"Yüklenen Resim: {image_path}")
print(f"Model Tahmini: **{predicted_class_name.upper()}**")
print(f"Güven Oranı: {confidence:.2f}%")

# 6. Çiçek Bilgisini Göster
display_flower_info(predicted_class_name)

# Resmi göstermek için
original_img = tf.io.read_file(image_path)
original_img = tf.image.decode_image(original_img, channels=3)

plt.figure(figsize=(8, 8))
plt.imshow(original_img.numpy())
plt.title(f"Tahmin: {predicted_class_name.capitalize()} ({confidence:.2f}%)", fontsize=16)
plt.axis('off')
plt.show()
```
